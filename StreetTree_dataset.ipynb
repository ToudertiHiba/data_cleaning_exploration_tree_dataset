# Data Cleaning

```python
# Replace missing values for certain conditions (columns for stumps and dead trees) with 'Not Applicable'
mask = ((tree_dataset_subset['status'] == 'Stump') | (tree_dataset_subset['status'] == 'Dead'))
tree_dataset_subset.loc[mask] = tree_dataset_subset.loc[mask].fillna('Not Applicable')
```

```python
# Handling missing values for health, sidewalk, species, and problems
tree_dataset_subset['problems'].fillna('None', inplace=True)
tree_dataset_subset['health'].fillna('Good', inplace=True)
tree_dataset_subset['spc_latin'].fillna('No Observation', inplace=True)
tree_dataset_subset['sidewalk'].fillna('NoDamage', inplace=True)
tree_dataset_subset['steward'].fillna('None', inplace=True)
```

```python
# Filter out small trees (with diameter less than or equal to 60)
small_trees = tree_dataset_subset[(tree_dataset_subset["tree_dbh"] <= 60) & (tree_dataset_subset["stump_diam"] <= 60)]
small_trees
```

In order to not take the outliers into consideration while computing the mean, we remove the trees with a diameter > 60 from the dataset.

```python
# Filter out alive small trees
small_trees_alive = small_trees[small_trees['status']=="Alive"]
# Filter out dead or stump small trees
small_trees_dead_or_stump = small_trees[(small_trees['status']=="Dead") | (small_trees['status']=="Stump")]
```

```python
# Display the mean 'tree_dbh' for each species among alive small trees
pd.DataFrame(small_trees_alive.groupby('spc_latin')['tree_dbh'].mean())
```

```python
# Display descriptive statistics for 'tree_dbh' for each species of alive small trees
small_trees_alive.groupby('spc_latin')['tree_dbh'].describe()
```

```python
# Create a table that shows the 25% and 75% percentiles for each species of alive small trees
stats_alive = small_trees_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive
```

```python
# Display rows with 'tree_dbh' equal to 0 among alive small trees
small_trees_alive[small_trees_alive['tree_dbh']==0]
```

We merge the table we just created (with the species and 25% and 75%) with the table containing all the alive small trees

```python
# Merge the table with species and percentiles with the table containing all alive small trees
small_trees_alive = small_trees_alive.merge(stats_alive, on='spc_latin', how='left')
small_trees_alive
```

We compare the diameter of each tree with the 25% and 75% values. If the diameter is lower than 25% value, we change it to 25%, if it is higher than 75%, we change it to 75%.

```python
# Compare the diameter of each tree with the 25% and 75% values.
# If the diameter is lower than the 25% value, set it to the 25% value.
mask25 = small_trees_alive['tree_dbh'] < small_trees_alive['25%']
small_trees_alive.loc[mask25, 'tree_dbh'] = small_trees_alive['25%']

# If the diameter is higher than the 75% value, set it to the 75% value.
mask75 = small_trees_alive['tree_dbh'] > small_trees_alive['75%']
small_trees_alive.loc[mask75, 'tree_dbh'] = small_trees_alive['75%']
```
